```python
# Create a sample MongoDB-style document
# This simulates a delivery record stored as a document

delivery_document = {
    "delivery_id": "D001",
    "delivery_status": "Delayed",
    "hub_id": "H05",
    "driver_id": "DR12",
    "vehicle_id": "VH07"
}

print("Sample Delivery Document:")
print(delivery_document)
```

```python
# Demonstrate CRUD operations conceptually

crud_operations = [
    "Create -> insert_one()",
    "Read -> find()",
    "Update -> update_one()",
    "Delete -> delete_one()"
]

print("MongoDB CRUD Operations:")
for operation in crud_operations:
    print(operation)
```

```python
# Example aggregation pipeline
# Used to summarise delivery outcomes

pipeline = [
    {
        "$group": {
            "_id": "$delivery_status",
            "count": {"$sum": 1}
        }
    }
]

print("Aggregation Pipeline Example:")
print(pipeline)
```
