```python
# Python Data Processing Notebook
# NorthStar Urban Mobility and Logistics Case Study

import pandas as pd
import numpy as np

print("Libraries imported successfully")
```

```python
# Create a sample deliveries dataset
# This simulates operational delivery records for analysis

data = {
    "delivery_id": ["D001", "D002", "D003", "D004"],
    "delivery_status": ["Delayed", "Completed", "Delayed", "Failed"],
    "hub_id": ["H01", "H02", "H01", "H03"]
}

df = pd.DataFrame(data)

print("Sample Dataset:")
print(df)
```

```python
# Check for missing values in the dataset

print("Missing Values:")
print(df.isnull().sum())
```

```python
# Count delivery outcomes
# Helps identify operational issues such as delays or failed deliveries

status_summary = df["delivery_status"].value_counts()

print("Delivery Status Summary:")
print(status_summary)
```

```python
# Convert delivery outcome counts into percentages

status_percentage = df["delivery_status"].value_counts(normalize=True) * 100

print("Delivery Outcome Mix (%):")
print(status_percentage)
```

```python
# Display basic dataset information

print("Dataset Information:")
print(df.info())
```
