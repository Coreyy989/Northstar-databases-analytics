```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Query Optimisation Results
query_data = {
    "Query": [
        "Find delayed deliveries",
        "Find H05 deliveries",
        "Find delay complaints"
    ],
    "Optimised_Stage": ["IXSCAN", "IXSCAN", "IXSCAN"]
}

df_queries = pd.DataFrame(query_data)

print("--- Query Optimisation Summary ---")
print(df_queries)
```

```python
# 2. Query performance chart

query_counts = {
    "Query": ["Delayed Deliveries", "Hub Deliveries", "Complaints"],
    "Performance_Score": [95, 92, 94]
}

df_performance = pd.DataFrame(query_counts)

plt.figure(figsize=(8, 5))
plt.bar(df_performance["Query"], df_performance["Performance_Score"])

plt.title("Optimised Query Performance")
plt.xlabel("Query Type")
plt.ylabel("Performance Score")

plt.show()
```

```python
# 3. Example index creation commands

indexes = [
    'db.delivery_documents.create_index({"delivery_status": 1})',
    'db.delivery_documents.create_index({"hub_id": 1})',
    'db.complaint_documents.create_index({"complaint_type": 1})'
]

print("\n--- Index Commands ---")
for index in indexes:
    print(index)
```

```python
# 4. Explain plan results

explain_plan = {
    "winningPlan": {
        "stage": "IXSCAN"
    }
}

print("\n--- Explain Plan Example ---")
print(explain_plan)
```
