```python
%%R

# Create sample complaint analytics dataset

complaints_data <- data.frame(
  complaint_type = c("Delay", "Damage", "Billing"),
  total = c(94, 30, 16)
)

print(complaints_data)
```

```python
%%R

# Summary statistics

summary(complaints_data)
```

```python
%%R

# Load visualisation library

library(ggplot2)
```

```python
%%R

# Create bar chart

ggplot(complaints_data, aes(x = complaint_type, y = total)) +
  geom_bar(stat = "identity") +
  theme_minimal() +
  labs(
    title = "Complaint Summary",
    x = "Complaint Type",
    y = "Count"
  )
```
